On va utiliser l'implémentation de ces classes de la manière suivante : 

```python
class LinkedList:

    def __init__(self):
        self.__length = 0
        self.__head = None

    def add(self, cargo):
        node = Node(cargo,self.__head)
        self.__head = node
        self.__length += 1

    def print(self):
        print("[", end=" ")
        if self.__head is not None:
            self.__head.print_list()
        print("]")

    #...

class Node:

    def __init__(self, cargo=None, next=None):
        self.__cargo = cargo
        self.__next  = next

    def value(self):
        return self.__cargo

    def __str__(self):
        return str(self.value())

    def print_list(self):
        print(self, end="")   # print my head
        tail = self.__next    # go to my next node
        if tail is not None : # as long as end of list was not reached
            tail.print_list() # recursively print remainder of list

    #...s
```

### Un peu de théorie et de vocabulaire 

> ### Qu'est ce qu'une liste chainée ? 

C'est une **structure de données** qui implémente une collection d'éléments sous la forme d'une **séquence de noeuds connectés**.

> ### Qu'est ce qu'une structure de donnée ? Donnez un exemple.

C'est un moyen de faciliter l'organisation et le regroupement de donner pour faciliter leur utilisation. Une liste de liens est un exemple typique de structure de données. Elle permet de regrouper des éléments en les organisant dans une structure que l'on peut parcourir de manière séquentielle.

> ### Expliquez la différence entre un noeud (Node) et sa cargaison (cargo).

Un noeud est un élément d'une liste chainée. Il s'agit d'un **objet** qui **contient une référence** intégrée **à un autre objet du même type** **ET** un élément de données appelé **sa cargaison**. La cargaison est composée des données contenues par un noeud. La référence est ce qui permet de relier les noeuds entre eux !

> ### Expliquez la différence entre un noeud (Node) et une liste chainée (LinkedList).

Le noeud sert à créer la liste chainée. Donc, une liste chainée est un ensemble de noeuds enchaînés. Il existe plusieurs types de liste chainée : doublement chainées, simplement chainées, cyclique, ...

> ### Dans le livre et le cours magistral, la notion d'une liste chaînée a d'abord été expliqué sans la classe LinkedList, simplement en chaînant des noeuds l'un à l'autre. Pourquoi avons nous alors besoin d'une classe LinkedList ?
